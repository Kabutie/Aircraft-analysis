```python
# Droping colums that have more missing values and i do not require for my analysis
aviation_df = aviation_df.drop(columns=['Air.carrier','Schedule','FAR.Description','Broad.phase.of.flight','Airport.Code','Airport.Name','Purpose.of.flight'])
aviation_df
```

```python
# Drop rows with missing aircraft type or fatalities
aviation_df = aviation_df.dropna(subset=['Make', 'Model', 'Total.Fatal.Injuries'])
aviation_df
```

```python
# Filter data to include accidents with valid event dates
aviation_df = aviation_df[aviation_df['Event.Date'].notnull()]
aviation_df.head()
```

```python
# visualizing the number of accidents over time
aviation_df['Event.Date'] = pd.to_datetime(aviation_df['Event.Date'], errors='coerce')
aviation_df['Year'] = aviation_df['Event.Date'].dt.year 
accidents_per_year = aviation_df.groupby('Year').size()

plt.figure(figsize=(10, 6))
accidents_per_year.plot(kind='line')
plt.title('Number of Aviation Accidents Over Time')
plt.xlabel('Year')
plt.ylabel('Accident Count')
plt.grid(True)
plt.tight_layout()
plt.show()

# Accidents merely happened in the past until around 1982 where the number of accidents increased.
```

```python
# Ensure injury columns are numeric
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']
aviation_df[injury_cols] = aviation_df[injury_cols].apply(pd.to_numeric, errors='coerce')

# Combine the columns
aviation_df['Total_Injuries'] = aviation_df[injury_cols].sum(axis=1)

# Group by Aircraft Category and sum total injuries
injuries_by_category = aviation_df.groupby('Aircraft.Category')['Total_Injuries'].sum().reset_index()

# Sort values for clearer plotting
injuries_by_category = injuries_by_category.sort_values(by='Total_Injuries', ascending=False)

# Visualize total injuries by aircraft categories
plt.figure(figsize=(10, 6))
sns.barplot(
    data=injuries_by_category,
    x='Total_Injuries',
    y='Aircraft.Category',

)
plt.title('Total Injuries by Aircraft Category')
plt.xlabel('Total Injuries')
plt.ylabel('Aircraft Category')
plt.tight_layout()
plt.show()
```

```python
# visualizing 
plt.figure(figsize=(14, 8))
sns.countplot(
    data=aviation_df,
    x='Aircraft.Category',
    hue='Severity.Category',
    palette='Set2'
)

plt.title('Aircraft Category by Categorized Injury Severity')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=60)
plt.legend(title='Severity Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
# Airplanes are high risk hence unsuitable for our company as it has the causes the most severe injuries
```

```python
# Replace 0.0 with NaN
aviation_df['Aircraft.Category'] = aviation_df['Aircraft.Category'].replace(0.0, np.nan)

# Drop missing categories for plotting
plot_df = aviation_df.dropna(subset=['Aircraft.Category'])

plt.figure(figsize=(10, 6))
ax = sns.countplot(
    data=plot_df,
    x='Aircraft.Category',
    order=plot_df['Aircraft.Category'].value_counts().index
)

# Axis labels and title
ax.set_title('Number of Accidents by Aircraft Category')
ax.set_xlabel('Aircraft Category')
ax.set_ylabel('Number of Accidents')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
#Airplanes are high risk for the company  since they have the highest number of accidents.
```

```python
# identifying the relationship between Makes, Models and engine types with accidents 
make_model_accident_counts = aviation_df.groupby(['Make', 'Model']).size().reset_index(name='AccidentCount', drop=False)
make_model_accident_counts = make_model_accident_counts.sort_values(by='AccidentCount', ascending=False)
make_model_accident_counts.head()

# The Cessna make has more accident counts compared to others hence the less suitable choice.
```

```python
# exploring engine type with more accidents occurences

aviation_df['Engine.Type'] = aviation_df['Engine.Type'].replace(0.0, np.nan) # some values in Engine column are 0.
# we don't have any aircraft with 0 engine. So I treat them as missing values

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=aviation_df, x='Engine.Type', order=aviation_df['Engine.Type'].value_counts().index)
ax.set_title('Aircraft Engine Types Involved in Accidents')
ax.set_ylabel('Number of Accidents')
ax.set_xlabel('Engine Type')
ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()

# Reciprocating engine type has the highest number of accidents hence we do not need that for our aircraft selection.
```

```python
# Group by Aircraft Category and Engine Type
category_engine_counts = aviation_df.groupby(['Aircraft.Category', 'Engine.Type']).size().reset_index(name='Count')

# Sort values for better readability
category_engine_counts = category_engine_counts.sort_values(by='Count', ascending=False)
print(category_engine_counts)
# Turbo shaft engine for Helicopter aircraft is more suitable since Helicopters have a lower number of accidents compared to airplanes and
#Turbo shaft has less accidents compared to the reciprocating engine type.
```

```python
# visualizing the number of accidents in different make and models of aircraft
# Combine 'Make' and 'Model' into a single column
aviation_df['Make_Model'] = aviation_df['Make'] + ' ' + aviation_df['Model']

# Count number of accidents per Make_Model
make_model_counts = aviation_df['Make_Model'].value_counts().reset_index()
make_model_counts.columns = ['Make_Model', 'AccidentCount']

# Get top 10
top_10_make_model = make_model_counts.head(10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top_10_make_model['Make_Model'], top_10_make_model['AccidentCount'])
ax.set_xlabel('Number of Accidents')
ax.set_ylabel('Make and Model')
ax.set_title('Top 10 Aircraft Make & Model by Accident Count')
plt.tight_layout()
plt.show()
# Cessna152, Cessna172, Cessna172N invoved in more accidents. This might be due to the populariy of these models
```

```python
#Identifying the most popular Helicopter by make and model.
#Filtering the DataFrame to include only rows where the aircraft category is 'Helicopter'
helicopters_df = aviation_df[aviation_df['Aircraft.Category'] == 'Helicopter']

# Count occurrences of each Make_Model
helicopter_counts = helicopters_df.groupby('Make_Model').size().reset_index(name='count')

# Sort descending for top 10 helicopter models
top_helicopters = helicopter_counts.sort_values(by='count', ascending=False).head(10)

# Print top 10 helicopter models
print("Top 10 Helicopter Make & Model by Number of Records:")
print(top_helicopters)

# visualization using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_helicopters,
    y='Make_Model',
    x='count',
    hue='Make_Model',
    palette='mako'
)
plt.title('Top 10 Helicopter Make & Model in Dataset')
plt.xlabel('Number of Records')
plt.ylabel('Make & Model')
plt.tight_layout()
plt.show()
#Robinson Helicopter company R44II is the most popular helicopter.
```

```python
# Filter to only helicopter accidents
helicopter_df = aviation_df[aviation_df['Aircraft.Category'] == 'Helicopter']

# Group by Make_Model and count accidents (each row = one accident)
helicopter_accident_counts = helicopter_df.groupby('Make_Model').size().reset_index(name='Accident_Count')

# Sort by accident count and select top 10
top_helicopter_accidents = helicopter_accident_counts.sort_values(by='Accident_Count', ascending=False).head(10)

# Print result
print("Top 10 Helicopters by Number of Accidents:")
print(top_helicopter_accidents)

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_helicopter_accidents,
    y='Make_Model',
    x='Accident_Count',
    hue='Make_Model',
    palette='rocket'
)
plt.title('Top 10 Helicopter Make & Model by Number of Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Make & Model')
plt.tight_layout()
plt.show()
#Robinson model Helicopter has the highest number of accidents hence a high risk choice for the Business.
```

```python

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=aviation_df, x='Amateur.Built', order=['No', 'Yes'])

# Add percentage labels
total = len(aviation_df)
for p in ax.patches:
    height = p.get_height()
    percent = f'{100 * height / total:.1f}%'
    ax.annotate(percent, (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom')

ax.set_title('Amateur Built Aircraft Involved in Accidents')
ax.set_ylabel('Number of Accidents')
ax.set_xlabel('Amateur Built')
plt.tight_layout()
plt.show()

# The aircrafts that have been built by amatuers that have been involved in accidents are few hence will not help in our analysis.
# They are not a suitable choice for the company to invest in though.
```
